# League Level: Logistic Regression IRL function

```python
# Extracting features and corresponding fourth down action (labels)
X = df_clean[candidate_features].values
y = df_clean['action'].values
    
# Fit multinomial logistic regression
irl_model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000)
irl_model.fit(X, y)
```

```python
# Interpret coefficients
print("FEATURE IMPORTANCE PER MODEL")
print("\n" + "*"*60)

for action_idx, action in enumerate(irl_model.classes_):
    print(f"\n{action.upper()}:")
    coefs = irl_model.coef_[action_idx]
        
    for feature, weight in zip(candidate_features, coefs):
        print(f"  {feature:20s}: {weight:+.4f}")
    
# Feature importance: Average weights across action
avg_weights = np.abs(irl_model.coef_).mean(axis=0)
print("\n" + "*"*60)
print("FEATURE IMPORTANCE (avg |weight| across actions):")
for feature, importance in sorted(zip(candidate_features, avg_weights),key=lambda x: x[1],reverse=True):
    print(f"  {feature:20s}: {importance:.4f}")
```

```python
# Accuracy Evaluation
print("MODEL PERFORMANCE METRICS EVALUATION")
print("*"*60)
    
y_pred = irl_model.predict(X)
# Overall accuracy
accuracy = (y_pred == y).mean()
print(f"\nOverall Accuracy: {accuracy:.1%}")
    
# Precision, Recall, F1 per action
precision, recall, f1, support = precision_recall_fscore_support( y, y_pred, labels=irl_model.classes_, zero_division=0)
    
print(f"\n{'Action':<10} {'Precision':<12} {'Recall':<12} {'F1-Score':<12} {'Support':<10}")
print("-" * 60)
for i, action in enumerate(irl_model.classes_):
    print(f"{action:<10} {precision[i]:<12.1%} {recall[i]:<12.1%} {f1[i]:<12.1%} {support[i]:<10.0f}")
    
# Macro and weighted averages
precision_macro = precision.mean()
recall_macro = recall.mean()
f1_macro = f1.mean()
    
precision_weighted = (precision * support).sum() / support.sum()
recall_weighted = (recall * support).sum() / support.sum()
f1_weighted = (f1 * support).sum() / support.sum()
    
print("-" * 60)
print(f"{'Macro Avg':<10} {precision_macro:<12.1%} {recall_macro:<12.1%} {f1_macro:<12.1%}")
print(f"{'Weighted Avg':<10} {precision_weighted:<12.1%} {recall_weighted:<12.1%} {f1_weighted:<12.1%}")
```

Understanding what IRL is doing:
For each action ('go', 'punt', 'fg'), the model computes a score:
    P(GO) / P(baseline) = exp(0.75  in_own_territory + ...)
The action with the highest score is what the model predicts the coach will choose.
Positive weight = higher feature value means more likely to choose this action
Negative weight = higher feature value means less likely to choose this action

Our model tells us that:

The most dominant factor in coach's fourth down decision are: desperation (must_score), distance to goal (very_short), having a good short-range kicker(fg_pct_short) and field position risk (in_own_territory).

R(action) = 1.7desperation + 1.6short_yardage + 1.0kicker_quality + 0.5offense_strength - 1.0field_position_risk

When desparate to score, teams are more likely to go for it and not punt or attempt field goal.
When trailing late and needing points, coaches abandon their usual conservatism. This makes sense - career risk doesn't matter if you're losing anyway.

For obvious reasons, when distance from goal is very short, teams are more likely to go for it and not punt or go for field goal. On 4th-and-1 or less, coaches trust their offense. This is one area where behavior aligns with analytics.

Good short-range kicker leads to more FG attempts and low likelihood of attempting punts.

In absolute terms, teams are LESS likely to go for it in own territory - they punt instead. But among the small number who don't punt, they're slightly more likely to go than kick FG (because FG from your own 30 is impossible). Coaches are obsessed with field position safety. When deep in their own territory, they punt overwhelmingly, even in situations where analytics might suggest going for it.

Protecting leads and recent offensive capabilities of team also become important. When ahead by one score late, coaches play not to lose. They are less likely to attempt going for it and more likely to punt. Good offense performance encourages coaches to go for it and not punt.

Shocking discoveries:
Weather is irrelevant and home crowd pressure doen't matter. Coaches don't factor special team skills beyond kickers much either.

Our findings show that coaching style remains conventional and rule-based in NFL.There is greater scope for analytics integration

Model Evaluation: 

88% overall accuracy with excellent FG (87% F1) and punt (95% F1) prediction. However, "go for it" recall remains 56% - the model misses 44% of aggressive decisions, suggesting significant unobserved factors (coach personality and intuition, momentum, player-specific matchups, player health).

# League model Evaluation

## Temporal Validation

```python
print("TEMPORAL VALIDATION")
print("*"*70)

# Split by season
train_mask = df_clean['season'] <= 2021
test_mask = df_clean['season'] >= 2022

X_train = df_clean.loc[train_mask, candidate_features].values
y_train = df_clean.loc[train_mask, 'action'].values
X_test = df_clean.loc[test_mask, candidate_features].values
y_test = df_clean.loc[test_mask, 'action'].values

print(f"Training set (2016-2021): {len(X_train)} decisions")
print(f"Test set (2022-2024):     {len(X_test)} decisions")
print(f"Total set:     {len(X_test)+len(X_train)} decisions (2016-2024)")
```

```python
# Train model
irl_temporal = LogisticRegression(
    multi_class='multinomial',
    solver='lbfgs',
    max_iter=1000,
    C=1.0
)
irl_temporal.fit(X_train, y_train)
```

```python
# Evaluate on test set
y_pred_test = irl_temporal.predict(X_test)
test_accuracy = (y_pred_test == y_test).mean()
train_accuracy = (irl_temporal.predict(X_train) == y_train).mean()

print(f"\nTraining accuracy (2016-2021): {train_accuracy:.1%}")
print(f"Test accuracy (2022-2024):     {test_accuracy:.1%}")
print(f"Generalization gap:            {(train_accuracy - test_accuracy)*100:.1f} pp")

# Detailed test set metrics
print("\n TEST SET PERFORMANCE (2022-2024)")
print("*"*70)
print(classification_report(y_test, y_pred_test, digits=3))

# Coefficient stability over time
print("\n COEFFICIENT STABILITY CHECK for GO action")

go_idx = list(irl_temporal.classes_).index('go')
go_idx_og = list(irl_model.classes_).index('go')
print(f"\n{'Feature':<30} {'2016-2021 Train':>15} {'2016-2024 Full':>15} {'Change':>15}")
print("*"*90)

for i, feature in enumerate(candidate_features):
    coef_train = irl_temporal.coef_[go_idx, i]
    coef_og = irl_model.coef_[go_idx, i]
    print(f"{feature:<30} {coef_train:>+15.4f} {coef_og:>+15.4f} {coef_og-coef_train:>+15.4f}")


# Coaching behavior changes over time
print("\n BEHAVIORAL CHANGES (2016-2021 vs 2022-2024)")
print("*"*70)

train_dist = pd.Series(y_train).value_counts(normalize=True).sort_index()
test_dist = pd.Series(y_test).value_counts(normalize=True).sort_index()

print(f"\n{'Action':<6} {'Train %':>10} {'Test %':>10} {'Change':>10}")
print("-"*45)
for action in ['fg', 'go', 'punt']:
    train_pct = train_dist.get(action, 0)
    test_pct = test_dist.get(action, 0)
    change = (test_pct - train_pct) * 100
    print(f"{action:<10} {train_pct:>9.1%} {test_pct:>9.1%} {change:>+9.1f} pp")

# Model stability and generalization over time
##threshold calc: Random sampling variation typically causes 1-2% fluctuation, 3% allows for some natural variance while indicating the model is stable.
##If train=88% and test=85-91%, With ~20,000 samples, SE  (0.880.12/20000)  0.002 (0.2%), 95% CI = 0.4%, so 3% is ~ 7 SE. 5% is industry std.
if abs(train_accuracy - test_accuracy) < 0.03:
    print("Coaching patterns stable over time - model generalizes well")
elif train_accuracy - test_accuracy < 0.05:
    print("Coaching may be evolving - model drift suspected")
else:
    print("Model performs well on future seasons")
```

# Team level Analysis - Go for it decisions

```python
# Small feature set to deal with 800-900 data rows per team
trunc_features = ['in_own_territory','very_short', 'must_score', 'score_differential', 'off_epa_4w']

# Get unique teams
teams = df_clean['posteam'].unique()
min_games = 200  # Need atleast 200 games per team

print(f"Analyzing teams with {min_games}+ games")
print(f"Features used {len(trunc_features)}")

team_results = {}

for team in teams:
    # Filter teams
    team_mask = (df_clean['posteam'] == team)
    n_games = team_mask.sum()
    
    if n_games >= min_games:
    
        # Prepare team-specific data
        X_team = df_clean.loc[team_mask, trunc_features].values
        y_team = df_clean.loc[team_mask, 'action'].values
    
        # Fit IRL model for each team
        team_model = LogisticRegression( solver='lbfgs', max_iter=1000, C=10.0)  # Higher regularization for small samples
        team_model.fit(X_team, y_team)
        
        # Predict and evaluate
        y_pred_team = team_model.predict(X_team)
        acc = (y_pred_team == y_team).mean()
        
        # Extract weights for 'go' action only (most interpretable)
        go_idx = list(team_model.classes_).index('go')
        team_weights = {}
        for i, feature in enumerate(trunc_features):
            team_weights[feature] = team_model.coef_[go_idx, i]
        
        # Calculate go rate for context
        go_rate = (y_team == 'go').mean()
        
        team_results[team] = {'weights': team_weights, 'n_games': n_games,'accuracy': acc, 'go_rate': go_rate,}

print(f"Successfully analyzed {len(team_results)} teams\n")
```

```python
# Create comparison dataframe
team_comparison = pd.DataFrame([
    {'team': team, **result['weights'], 
     'n_games': result['n_games'],
     'accuracy': result['accuracy'],
     'go_rate': result['go_rate']}
    for team, result in team_results.items()
])

# Display results
print("TEAM DECISION-MAKING PATTERNS")
print("*"*70)

print("\n1. Top-5 Aggressive (Go for it) teams :")
print(team_comparison.nlargest(5, 'go_rate')[
    ['team', 'go_rate', 'must_score', 'very_short', 'n_games']
].to_string(index=False))

print("\n2. Top-5 Teams concerned by field-position safety:")
print(team_comparison.nsmallest(5, 'in_own_territory')[
    ['team', 'in_own_territory', 'go_rate', 'n_games']
].to_string(index=False))

print("\n3. Top-5 Teams driven by desperation:")
print(team_comparison.nlargest(5, 'must_score')[
    ['team', 'must_score', 'go_rate', 'n_games']
].to_string(index=False))

print("\n4. Top-5 Teams with strong recent offense performance:")
print(team_comparison.nlargest(5, 'off_epa_4w')[
    ['team', 'off_epa_4w', 'go_rate', 'n_games']
].to_string(index=False))

# Team similarity
print("\nHETEROGENEITY ACROSS TEAMS \n")

for feature in trunc_features:
    mean_w = team_comparison[feature].mean()
    std_w = team_comparison[feature].std()
    print(f"{feature:25s}: mean={mean_w:+.3f}, std={std_w:.3f}")

# Test for meaningful variation
std_threshold = 0.5
high_variation_features = [
    f for f in trunc_features 
    if team_comparison[f].std() > std_threshold
]

# Summary statistics
print("\nSUMMARY \n")
print(f"Average go rate: {team_comparison['go_rate'].mean():.1%}")
print(f"Go rate range: {team_comparison['go_rate'].min():.1%} - {team_comparison['go_rate'].max():.1%}")
print(f"Average accuracy: {team_comparison['accuracy'].mean():.1%}")
```
