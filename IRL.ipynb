# League Level: Logistic Regression IRL function

```python
def simple_irl(df, reward_features):
    """
    Simpler IRL: Find weights that best predict which action was chosen.
    
    For each play, compute R(s,a) = ^T (s,a) for all actions.
    Learn  such that actual action has highest reward.
    """    
    
    # Extracting features and corresponding fourth down action (labels)
    X = df[reward_features].values
    y = df['action'].values
    
    # Fit multinomial logistic regression
    model = LogisticRegression(
        multi_class='multinomial',
        solver='lbfgs',
        max_iter=1000
    )
    model.fit(X, y)
    
    # Interpret coefficients
    print("\n" + "*"*60)
    print("INFERRING CANDIDATE REWARD FEATURE WEIGHTS")
    print("*"*60)
    
    for action_idx, action in enumerate(model.classes_):
        print(f"\n{action.upper()}:")
        coefs = model.coef_[action_idx]
        
        for feature, weight in zip(reward_features, coefs):
            print(f"  {feature:20s}: {weight:+.4f}")
    
    # Average absolute weight = importance
    avg_weights = np.abs(model.coef_).mean(axis=0)
    print("\n" + "*"*60)
    print("FEATURE IMPORTANCE (avg |weight| across actions):")
    for feature, importance in sorted(zip(reward_features, avg_weights),
                                     key=lambda x: x[1],
                                     reverse=True):
        print(f"  {feature:20s}: {importance:.4f}")
 
    # Accuracy Evaluation
    print("\n" + "*"*60)
    print("MODEL PERFORMANCE METRICS EVALUATION")
    print("*"*60)
    
    y_pred = model.predict(X)
    # Overall accuracy
    accuracy = (y_pred == y).mean()
    print(f"\nOverall Accuracy: {accuracy:.1%}")
    
    # Precision, Recall, F1 per action
    precision, recall, f1, support = precision_recall_fscore_support(
        y, y_pred, labels=model.classes_, zero_division=0
    )
    
    print(f"\n{'Action':<10} {'Precision':<12} {'Recall':<12} {'F1-Score':<12} {'Support':<10}")
    print("-" * 60)
    for i, action in enumerate(model.classes_):
        print(f"{action:<10} {precision[i]:<12.1%} {recall[i]:<12.1%} {f1[i]:<12.1%} {support[i]:<10.0f}")
    
    # Macro and weighted averages
    precision_macro = precision.mean()
    recall_macro = recall.mean()
    f1_macro = f1.mean()
    
    precision_weighted = (precision * support).sum() / support.sum()
    recall_weighted = (recall * support).sum() / support.sum()
    f1_weighted = (f1 * support).sum() / support.sum()
    
    print("-" * 60)
    print(f"{'Macro Avg':<10} {precision_macro:<12.1%} {recall_macro:<12.1%} {f1_macro:<12.1%}")
    print(f"{'Weighted Avg':<10} {precision_weighted:<12.1%} {recall_weighted:<12.1%} {f1_weighted:<12.1%}")
       
    return model
```

```python
# Run simple IRL
irl_model = simple_irl(df_clean, candidate_features)
```

Understanding what IRL is doing:
For each action ('go', 'punt', 'fg'), the model computes a score:
    P(GO) / P(baseline) = exp(0.75  in_own_territory + ...)
The action with the highest score is what the model predicts the coach will choose.
Positive weight = higher feature value means more likely to choose this action
Negative weight = higher feature value means less likely to choose this action

Our model tells us that:

The most dominant factor in coach's fourth down decision are: desperation (must_score), distance to goal (very_short), having a good short-range kicker(fg_pct_short) and field position risk (in_own_territory).

R(action) = 1.7desperation + 1.6short_yardage + 1.0kicker_quality + 0.5offense_strength - 1.0field_position_risk

When desparate to score, teams are more likely to go for it and not punt or attempt field goal.
When trailing late and needing points, coaches abandon their usual conservatism. This makes sense - career risk doesn't matter if you're losing anyway.

For obvious reasons, when distance from goal is very short, teams are more likely to go for it and not punt or go for field goal. On 4th-and-1 or less, coaches trust their offense. This is one area where behavior aligns with analytics.

Good short-range kicker leads to more FG attempts and low likelihood of attempting punts.

In absolute terms, teams are LESS likely to go for it in own territory - they punt instead. But among the small number who don't punt, they're slightly more likely to go than kick FG (because FG from your own 30 is impossible). Coaches are obsessed with field position safety. When deep in their own territory, they punt overwhelmingly, even in situations where analytics might suggest going for it.

Protecting leads and recent offensive capabilities of team also become important. When ahead by one score late, coaches play not to lose. They are less likely to attempt going for it and more likely to punt. Good offense performance encourages coaches to go for it and not punt.

Shocking discoveries:
Weather is irrelevant and home crowd pressure doen't matter. Coaches don't factor special team skills beyond kickers much either.

Our findings show that coaching style remains conventional and rule-based in NFL.There is greater scope for analytics integration

Model Evaluation: 

88% overall accuracy with excellent FG (87% F1) and punt (95% F1) prediction. However, "go for it" recall remains 56% - the model misses 44% of aggressive decisions, suggesting significant unobserved factors (coach personality and intuition, momentum, player-specific matchups, player health).

## Temporal Validation

```python
print("TEMPORAL VALIDATION: Train on 2016-2021, Test on 2022-2024")
print("="*70)

# Split by season
train_mask = df_clean['season'] <= 2021
test_mask = df_clean['season'] >= 2022

X_train = df_clean.loc[train_mask, candidate_features].values
y_train = df_clean.loc[train_mask, 'action'].values
X_test = df_clean.loc[test_mask, candidate_features].values
y_test = df_clean.loc[test_mask, 'action'].values

print(f"\nTraining set: {len(X_train)} decisions (2016-2021)")
print(f"Test set:     {len(X_test)} decisions (2022-2024)")
```

```python
# Train model
model_temporal = LogisticRegression(
    multi_class='multinomial',
    solver='lbfgs',
    max_iter=1000,
    C=1.0
)
model_temporal.fit(X_train, y_train)
```

```python
# Evaluate on test set
y_pred_test = model_temporal.predict(X_test)
test_accuracy = (y_pred_test == y_test).mean()
train_accuracy = (model_temporal.predict(X_train) == y_train).mean()

print(f"\nTraining accuracy (2016-2021): {train_accuracy:.1%}")
print(f"Test accuracy (2022-2024):     {test_accuracy:.1%}")
print(f"Generalization gap:            {(train_accuracy - test_accuracy)*100:.1f} pp")

# Detailed test set metrics
print("\n TEST SET PERFORMANCE (2022-2024)")
print("*"*70)
print(classification_report(y_test, y_pred_test, digits=3))

# Compare coefficients - are they stable over time?
print("\n COEFFICIENT STABILITY CHECK (GO action)")
print("*"*70)

go_idx = list(model_temporal.classes_).index('go')
print(f"{'Feature':<30} {'2016-2021 Train':>15} {'Direction':>10}")
print("-"*60)

for i, feature in enumerate(candidate_features):
    coef = model_temporal.coef_[go_idx, i]
    direction = "positive" if coef > 0 else "negative"
    print(f"{feature:<30} {coef:>+15.4f} {direction:>10}")

# Check if coaching behavior changed over time
print("\n BEHAVIORAL CHANGES (2016-2021 vs 2022-2024)")
print("*"*70)

train_dist = pd.Series(y_train).value_counts(normalize=True).sort_index()
test_dist = pd.Series(y_test).value_counts(normalize=True).sort_index()

print(f"\n{'Action':<10} {'Train %':>10} {'Test %':>10} {'Change':>10}")
print("-"*45)
for action in ['fg', 'go', 'punt']:
    train_pct = train_dist.get(action, 0)
    test_pct = test_dist.get(action, 0)
    change = (test_pct - train_pct) * 100
    print(f"{action:<10} {train_pct:>9.1%} {test_pct:>9.1%} {change:>+9.1f} pp")

# Model's ability to predict future behavior
##threshold calc: Random sampling variation typically causes 1-2% fluctuation, 3% allows for some natural variance while indicating the model is stable.
##If train=88% and test=85-91%, With ~20,000 samples, SE  (0.880.12/20000)  0.002 (0.2%), 95% CI = 0.4%, so 3% is ~ 7 SE. 5% is industry std.
if abs(train_accuracy - test_accuracy) < 0.03:
    print("\n Model generalizes well - coaching patterns stable over time")
elif train_accuracy - test_accuracy < 0.05:
    print("\n Significant performance drop - coaching may be evolving")
else:
    print("\n Model performs well on future seasons")
```

## Qualitative inspection

```python
# Create evaluation dataframe
df_eval = df_clean.copy()
df_eval['predicted'] = irl_model.predict(df_clean[candidate_features].values)
df_eval['correct'] = (df_eval['predicted'] == df_eval['action'])

# Get prediction probabilities
y_proba = irl_model.predict_proba(df_clean[candidate_features].values)
df_eval['pred_confidence'] = y_proba.max(axis=1)

# 1. High-confidence errors (model was sure but wrong)
print("*"*70)
print("\n1. HIGH-CONFIDENCE ERRORS (Model >80% sure but wrong)")
print("*"*70)

##logic for identifying confident errors
confident_errors = df_eval[
    (~df_eval['correct']) & (df_eval['pred_confidence'] > 0.8)
].sort_values('pred_confidence', ascending=False)

## confident errors reporting
print(f"Found {len(confident_errors)} high-confidence errors ({len(confident_errors)/len(df_eval)*100:.1f}% of data)\n")
print(confident_errors.head(10)[[
    'season', 'week', 'posteam', 'yardline_100', 'ydstogo', 
    'score_differential', 'must_score', 'action', 'predicted', 'pred_confidence']].to_string(index=False))

# 2. Systematic patterns in errors
print("*"*70)
print("\n2. ERROR PATTERNS BY SITUATION")
print("*"*70)

## Errors by action type
print("\nErrors by actual action:")
for action in ['fg', 'go', 'punt']:
    action_errors = df_eval[
        (df_eval['action'] == action) & (~df_eval['correct'])
    ]
    error_rate = len(action_errors) / (df_eval['action'] == action).sum()
    
    if len(action_errors) > 0:
        # Model predictions
        confusion = action_errors['predicted'].value_counts()
        print(f"\n  {action.upper()}: {error_rate:.1%} error rate")
        print(f"    Confused with: {confusion.to_dict()}")

# 3. Specific problematic contexts
print("*"*70)
print("\n3. PROBLEMATIC CONTEXTS")
print("*"*70)

# Context 1: Short yardage in own territory
pc_1 = df_eval[
    (df_eval['very_short'] == 1) & 
    (df_eval['in_own_territory'] == 1)
]
if len(pc_1) > 0:
    acc_1 = pc_1['correct'].mean()
    print(f"\nShort yardage in own territory: {acc_1:.1%} accuracy")
    print(f"  Sample size: {len(pc_1)}")
    if acc_1 < 0.7:
        print(" Model struggles with competing pressures (achievable vs risky)")

# Context 2: Must-score situations
pc_2 = df_eval[df_eval['must_score'] == 1]
if len(pc_2) > 0:
    acc_2 = pc_2['correct'].mean()
    print(f"\nMust-score situations: {acc_2:.1%} accuracy")
    print(f"  Sample size: {len(pc_2)}")
    if acc_2 < 0.7:
        print("Model struggles with desperation logic")

# Context 3: Marginal FG range (45-55 yards)
pc_3 = df_eval[
    (df_eval['yardline_100'] >= 28) & 
    (df_eval['yardline_100'] <= 38)
]
if len(pc_3) > 0:
    acc_3 = pc_3['correct'].mean()
    print(f"\nMarginal FG range (45-55 yards): {acc_3:.1%} accuracy")
    print(f"  Sample size: {len(pc_3)}")
    if acc_3 < 0.8:
        print("FG/punt decision boundary is fuzzy")

# 4. Cases where GO was correct but model predicted conservatively
print("*"*70)
print("\n4. IGNORED AGGRESSIVE OPPORTUNITIES")
print("*"*70)

missed_go = df_eval[
    (df_eval['action'] == 'go') & 
    (df_eval['predicted'] != 'go')
].sort_values('pred_confidence', ascending=False)

print(f"\nModel missed {len(missed_go)} 'go' decisions ({len(missed_go)/(df_eval['action']=='go').sum():.1%} of all go's)\n")
print("Top 10 most confident conservative predictions (but coach went for it):")
print(missed_go.head(10)[[
    'season', 'posteam', 'yardline_100', 'ydstogo', 'score_differential',
    'must_score', 'predicted', 'pred_confidence'
]].to_string(index=False))

# 5. Summary insights
print("*"*70)
print("\n. SUMMARY: What the Model Can't Capture")
print("*"*70)

# Calculate error rates by feature values
print("\nError rates suggest unobserved factors:")

# High offense EPA but still punted (trusting defense instead?)
conservative_despite_strength = df_eval[
    (df_eval['off_epa_4w'] > 0.2) & 
    (df_eval['action'] == 'punt') &
    (df_eval['very_short'] == 1)
]
if len(conservative_despite_strength) > 10:
    print(f"  - {len(conservative_despite_strength)} cases: Strong offense + short yardage  still punted")
    print("     Suggests: Defensive matchups, game script, or coach philosophy")

# Went for it despite risky context
aggressive_despite_risk = df_eval[
    (df_eval['in_own_territory'] == 1) &
    (df_eval['action'] == 'go') &
    (df_eval['very_short'] == 0)
]
if len(aggressive_despite_risk) > 10:
    print(f"  - {len(aggressive_despite_risk)} cases: Own territory + not short yardage  still went for it")
    print("     Suggests: Momentum, opponent weakness, or specific play confidence")
```

# Team level Analysis - all fourth down decisions

```python
# Minimal feature set (5 features for stability with limited data)
team_irl_features = [
    'in_own_territory',       # Field position risk
    'very_short',             # Short yardage
    'must_score',             # Desperation
    'score_differential',     # Game situation
    'off_epa_4w',            # Offensive strength
]

# Get unique teams
teams = df_clean['posteam'].unique()
min_decisions = 200  # Minimum threshold for reliable estimates

print(f"\nAnalyzing teams with {min_decisions}+ decisions")
print(f"Using {len(team_irl_features)} features\n")

team_results = {}

for team in teams:
    # Filter to this team's decisions
    team_mask = (df_clean['posteam'] == team)
    n_decisions = team_mask.sum()
    
    if n_decisions < min_decisions:
        continue
    # Prepare team-specific data
    X_team = df_clean.loc[team_mask, team_irl_features].values
    y_team = df_clean.loc[team_mask, 'action'].values
    
    # Fit team-specific IRL model
    team_model = LogisticRegression(
        multi_class='multinomial',
        solver='lbfgs',
        max_iter=1000,
        C=10.0  # Higher regularization for small samples
    )
    
    try:
        team_model.fit(X_team, y_team)
        
        # Predict and evaluate
        y_pred_team = team_model.predict(X_team)
        acc = (y_pred_team == y_team).mean()
        
        # Extract weights for 'go' action only (most interpretable)
        go_idx = list(team_model.classes_).index('go')
        team_weights = {}
        for i, feature in enumerate(team_irl_features):
            team_weights[feature] = team_model.coef_[go_idx, i]
        
        # Calculate go rate for context
        go_rate = (y_team == 'go').mean()
        
        team_results[team] = {
            'weights': team_weights,
            'n_decisions': n_decisions,
            'accuracy': acc,
            'go_rate': go_rate,
        }
        
    except Exception as e:
        print(f"Skipping {team}: {e}")
        continue

print(f"Successfully analyzed {len(team_results)} teams\n")
```

```python
# Create comparison dataframe
team_comparison = pd.DataFrame([
    {'team': team, **result['weights'], 
     'n_decisions': result['n_decisions'],
     'accuracy': result['accuracy'],
     'go_rate': result['go_rate']}
    for team, result in team_results.items()
])

# Display results
print("="*70)
print("TEAM DECISION-MAKING PATTERNS")
print("="*70)

print("\n1. Most AGGRESSIVE teams (highest go-for-it rate):")
print(team_comparison.nlargest(5, 'go_rate')[
    ['team', 'go_rate', 'must_score', 'very_short', 'n_decisions']
].to_string(index=False))

print("\n2. Most FIELD-POSITION SENSITIVE (large negative in_own_territory):")
print(team_comparison.nsmallest(5, 'in_own_territory')[
    ['team', 'in_own_territory', 'go_rate', 'n_decisions']
].to_string(index=False))

print("\n3. Most DESPERATION-DRIVEN (high must_score weight):")
print(team_comparison.nlargest(5, 'must_score')[
    ['team', 'must_score', 'go_rate', 'n_decisions']
].to_string(index=False))

print("\n4. Most OFFENSE-CONFIDENT (high off_epa_4w weight):")
print(team_comparison.nlargest(5, 'off_epa_4w')[
    ['team', 'off_epa_4w', 'go_rate', 'n_decisions']
].to_string(index=False))

# Variation analysis
print("\n" + "="*70)
print("HETEROGENEITY ACROSS TEAMS")
print("="*70)

for feature in team_irl_features:
    mean_w = team_comparison[feature].mean()
    std_w = team_comparison[feature].std()
    print(f"{feature:25s}: mean={mean_w:+.3f}, std={std_w:.3f}")

# Test for meaningful variation
std_threshold = 0.5
high_variation_features = [
    f for f in team_irl_features 
    if team_comparison[f].std() > std_threshold
]

print("\n" + "-"*70)
if high_variation_features:
    print(f"Features with substantial variation (std > {std_threshold}):")
    for feature in high_variation_features:
        print(f"  - {feature}")
    print("\n Some team-level heterogeneity exists")
else:
    print(" LIMITED variation across teams")
    print("   League-wide consensus on 4th down decision-making")

# Summary statistics
print("\n" + "="*70)
print("SUMMARY")
print("="*70)
print(f"Teams analyzed: {len(team_results)}")
print(f"Average go rate: {team_comparison['go_rate'].mean():.1%}")
print(f"Go rate range: {team_comparison['go_rate'].min():.1%} - {team_comparison['go_rate'].max():.1%}")
print(f"Average accuracy: {team_comparison['accuracy'].mean():.1%}")
```
